```python
#load csv file using pandas library function read_csv
import pandas as pd
data = pd.read_csv("online_retail.csv")
```

```python
data.fillna(0,inplace=True)
```

```python
#drop duplicate data
filtered_data=data[['Country','CustomerID']].drop_duplicates()
```

```python
#you can observe most of the customers are from "United Kingdom". So, you can filter data for United Kingdom customer.
uk_data=data[data.Country=='United Kingdom']
uk_data.info()
```

```python
#use to describe the data
uk_data.describe()
```

```python
uk_data=uk_data[['CustomerID','InvoiceDate','InvoiceNo','Quantity','UnitPrice']]
```

```python
#Calulate total purchase
uk_data['TotalPurchase'] = uk_data['Quantity'] * uk_data['UnitPrice']
```

```python
uk_data['InvoiceDate'] = pd.to_datetime(uk_data['InvoiceDate'])

# Grouping by 'CustomerID' and aggregating various metrics
uk_data_group = uk_data.groupby('CustomerID').agg({
    'InvoiceDate': lambda date: (date.max() - date.min()).days,
    'InvoiceNo': lambda num: len(num),
    'Quantity': lambda quant: quant.sum(),
    'TotalPurchase': lambda price: price.sum()
})
```

```python
# Change the name of columns
uk_data_group.columns=['num_days','num_transactions','num_units','spent_money']
uk_data_group.head()
```

```python
#Calculate CLTV using following formula:
```

```python
#1. Calculate Average Order Value
# Average Order Value
uk_data_group['avg_order_value']=uk_data_group['spent_money']/uk_data_group['num_transactions']
```

```python
uk_data_group.head()
```

```python
#Calculating Purchase Frequency
#By dividing the total number of transactions by the number of customers, calculating the average number of transactions per customer, which can be interpreted as the purchase frequency.
purchase_frequency=sum(uk_data_group['num_transactions'])/uk_data_group.shape[0]
purchase_frequency
```

```python
# Repeat Rate
#calculating the repeat rate, which is the proportion of customers who have made more than one purchase (transactions) to the total number of customers in the   dataset.
repeat_rate=uk_data_group[uk_data_group.num_transactions > 1].shape[0]/uk_data_group.shape[0]
repeat_rate
```

```python
#Churn Rate
#calculates the portion of customers who made only one purchase.
churn_rate=1-repeat_rate
```

```python
#Calculate Profit Margin
```

```python
# Profit Margin
uk_data_group['profit_margin']=uk_data_group['spent_money']*0.05
uk_data_group.head()

```

```python
#1. Calculate Average Order Value
# Average Order Value
uk_data_group['avg_order_value']=uk_data_group['spent_money']/uk_data_group['num_transactions']
```

```python
# Customer Value
uk_data_group['CLV']=(uk_data_group['avg_order_value']*purchase_frequency)/churn_rate
```

```python
#Customer Lifetime Value
uk_data_group['cust_lifetime_value']=uk_data_group['CLV']*uk_data_group['profit_margin']
```

Prediction Model for CLTV(Customer Life Time Value)

```python
#creating a new column named 'month_yr' in a DataFrame 'uk_data' to extract the month and year from the 'InvoiceDate' column and format it as 'MMM-YYYY'
uk_data['month_yr'] = uk_data['InvoiceDate'].apply(lambda x: x.strftime('%b-%Y'))
uk_data.head()
```

```python
#pivot_table function from pandas to create a table that summarizes the total purchase amount for each customer across different months and years. This table is pivoted in a way that rows represent individual customers ('CustomerID'),
#columns represent months and years ('month_yr'), and the values in the table are the total purchase amounts.
sale=uk_data.pivot_table(index=['CustomerID'],columns=['month_yr'],values='TotalPurchase',aggfunc='sum',fill_value=0).reset_index()
```

```python
sale.head()
```

```python
#adding up the values across columns starting from the third column (index position 2) to the last column for each row in a DataFrame named 'sale'.
#After calculating this sum, it assigns these values as a new column named 'CLV' (Customer Lifetime Value) in the DataFrame 'sale'.
sale['CLV']=sale.iloc[:,2:].sum(axis=1)
sale.head()
```

```python
X=sale[['Dec-2011','Nov-2011', 'Oct-2011','Sep-2011','Aug-2011','Jul-2011']]
y=sale[['CLV']]
#split training set and test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,random_state=0)
# import model
from sklearn.linear_model import LinearRegression

# instantiate-ABLE TO CREATE MODEL
linreg = LinearRegression()

# fit the model to the training data (learn the coefficients)
linreg.fit(X_train, y_train)

# make predictions on the testing set
y_pred = linreg.predict(X_test)
```

```python
# print the intercept and coefficients
print(linreg.intercept_)
print(linreg.coef_)
```

```python
from sklearn import metrics

# compute the R Square for model
print("R-Square:",metrics.r2_score(y_test, y_pred))
```

```python
# calculate MAE using scikit-learn
print("MAE:",metrics.mean_absolute_error(y_test,y_pred))

#calculate mean squared error
print("MSE",metrics.mean_squared_error(y_test, y_pred))
# compute the RMSE of our predictions
import numpy as np
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
